```python
import numpy as np
import matplotlib.pyplot as plt

xDomain = [ -2,2 ]

# option 1
resolution = .1
x = np.arange(xDomain[0], xDomain[1]+resolution, resolution)

# function
y = x**2 + 3*x**3 - x**4

#plot
plt.plot(x,y,label='$y=x^2+x^3-x^4$')
plt.grid()
plt.legend()
plt.xlim(xDomain[0],xDomain[1])
plt.xlabel('x')
plt.ylabel('y=f(x)')
plt.show()
```
